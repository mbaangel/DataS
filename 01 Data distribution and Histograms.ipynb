```python
import numpy as np

np.random.rand(10)
np.random.normal(loc=5, scale=10, size=1000).std()
```

Three important distributions - Normal, Uniform and Exponential

```python
data = np.random.normal(size=1000000)
plt.hist(data, color="pink", ec="black", alpha=0.5)
plt.title("Normal Distribution")
plt.show()

data = np.random.normal(size=1000000)
plt.hist(data, color="skyblue", ec="black", alpha=0.5)
plt.title("Uniform Distribution")
plt.show()

data = np.random.exponential(size=1000000)
plt.hist(data, color="green", ec="black", alpha=0.5)
plt.title("Exponential Distribution")
plt.show()
```
