```python
import numpy as np
import matplotlib.pyplot as plt

ages = np.array([4, 17, 29, 33, 50, 51, 61, 71, 85, 92, 99])
plt.yticks(np.arange(0,110, 5))
plt.grid(True)
plt.boxplot(ages)
plt.show()
```

```python
(np.min(ages), np.quantile(ages, 0.25), np.median(ages), np.quantile(ages, 0.75), np.max(ages))
```
